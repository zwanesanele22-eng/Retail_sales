```sql
SELECT*
FROM brightlearn.data.retail_sales_database;

------How many total transactions are recorded in this dataset?
SELECT COUNT(*) AS total_transactions
FROM  brightlearn.data.retail_sales_database;

-----------What is the total revenue generated by each product category
SELECT `Product Category`, SUM(`Total Amount`) AS total_revenue
FROM    brightlearn.data.retail_sales_database
GROUP BY `Product Category`;

------------List all transactions where the Product Category is "Electronics"
SELECT*
FROM   brightlearn.data.retail_sales_database
WHERE  `Product Category`= 'Electronics';

-------Which product categories had a total revenue greater than 50,000?
SELECT `Product Category`, SUM(`Total Amount`) AS total_revenue
FROM brightlearn.data.retail_sales_database
GROUP BY `Product Category`
HAVING SUM(`Total Amount`) >50000;

-----Find all transactions where the customer's age is between 25 and 40(inclusive)
SELECT*
FROM brightlearn.data.retail_sales_database
WHERE Age BETWEEN 25 AND 40;

-------Retrieve all transactions for the Product Categories 'Beauty'and 'Clothing' using IN
SELECT*
FROM   brightlearn.data.retail_sales_database
WHERE `Product Category`IN ('Beauty' , 'Clothing');

-----Find all transactions where Gender is "Female" and Total Amount is greater than 500
SELECT*
FROM  brightlearn.data.retail_sales_database
WHERE Gender='Female' AND `Total Amount`>500;

------Find electronics transactions where Total Amount is between 500 and 2000, and the customer is Male or aged over 50
SELECT*
FROM  brightlearn.data.retail_sales_database
WHERE `Product Category`='Electronics' AND `Total Amount`BETWEEN 500 AND 2000
       AND (Gender= 'Male' OR Age> 50);










```
